# Analyse exploratoire des données e-commerce
Analyse du dataset nettoyé : distributions, tendances, valeurs manquantes.

```
import sys; sys.path.append('..')
import pandas as pd
import plotly.express as px
df = pd.read_csv('../data/processed/dataset_clean.csv', parse_dates=['date'])
df.head()
```

```
df.info()
df.describe()
```

## CA par mois (tendance / saisonnalité)

```
ca_mois = df.groupby('mois')['montant'].sum().reset_index()
px.line(ca_mois, x='mois', y='montant', markers=True, title='CA par mois')
```

## Top produits

```
top = df.groupby('produit')['quantite'].sum().sort_values(ascending=False).head(10)
top
```
